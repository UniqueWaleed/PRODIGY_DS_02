Perform data cleaning and exploratory data on a data set of your choice, such as the titanic data set from kaggle. explore the relationships between variable and identify patterns and trends in the data set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

titanic_data = pd.read_csv("/content/train.csv")

print(titanic_data.head())
print(titanic_data.info())
print(titanic_data.describe())
```

```python
missing_values = titanic_data.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(titanic_data["Age"], bins=20, kde=True, color='skyblue')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Distribution of Passenger Ages")
plt.show()
```

**Relationship Between Gender (Sex) and Survival:**



```python
sns.barplot(x="Sex", y="Survived", data=titanic_data, errcolor=None)
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender")
plt.show()
```

**Relationship Between Sibling/Spouse Count (SibSp) and Survival:**



```python
sns.barplot(x="SibSp", y="Survived", data=titanic_data, errcolor=None)
plt.xlabel("Number of Siblings/Spouses")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Sibling/Spouse Count")
plt.show()
```

**Relationship Between Passenger Class (Pclass) and Survival:**



```python
sns.barplot(x="Pclass", y="Survived", data=titanic_data, errcolor=None)
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Passenger Class")
plt.show()
```

**Relationship Between Parent/Child Count (Parch) and Survival:**



```python
sns.barplot(x="Parch", y="Survived", data=titanic_data, errcolor=None)
plt.xlabel("Number of Parents/Children")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Parent/Child Count")
plt.show()
```
